```python
import requests
from bs4 import BeautifulSoup
import selenium


def get_html(url):
    response = requests.get(url)
    return response.text


def get_soup(html):
    return BeautifulSoup(html, "html.parser")


def get_data(url):
    html = get_html(url)
    soup = get_soup(html)
    return soup
```
